```python
#1
# Основные отличия между NLTK и SpaCy токенизаторами:

# - NLTK - более гибкий и настраиваемый токенизатор, который позволяет изменять его поведение в зависимости от требований задачи.
# - SpaCy - более современный и эффективный токенизатор с более точной логикой разбиения текста на токены.

# Достоинства NLTK:
# - Гибкость и настраиваемость.
# - Большое сообщество пользователей и разработчиков, что обеспечивает хорошую поддержку и доступность дополнительных ресурсов.

# Недостатки NLTK:
# - Менее эффективная скорость работы по сравнению с более современными библиотеками.

# Достоинства SpaCy:
# - Более высокая скорость работы.
# - Более точные результаты токенизации.

# Недостатки SpaCy:
# - Менее гибкий и настраиваемый, чем NLTK.

import spacy
from nltk import WhitespaceTokenizer

nlp = spacy.load("en_core_web_sm")

with open(
    "C:\\Users\\user\\Desktop\\python\\Tokenizer\\1.txt", encoding="utf-8"
) as file:
    text = file.read()
    doc = nlp(text)
    print("Noun phrases:", [chunk.text for chunk in doc.noun_chunks])
    print("Verbs:", [token.lemma_ for token in doc if token.pos_ == "VERB"],end="/n")

with open(
    "C:\\Users\\user\\Desktop\\python\\Tokenizer\\1.txt", encoding="utf-8"
) as file:
    text = file.read()
    tk = WhitespaceTokenizer()
    print(tk.tokenize(text))
```

```python
#2
import nltk
import string

nltk.download("punkt")

with open(
    "C:\\Users\\user\\Desktop\\python\\Tokenizer\\1.txt", encoding="utf-8"
) as file:
    text = file.read()
    tokens = nltk.word_tokenize(text)
    remove_punctuation = str.maketrans("", "", string.punctuation)
    tokens_ = [
        x
        for x in [t.translate(remove_punctuation).lower() for t in tokens]
        if len(x) > 0
    ]
    text2 = nltk.Text(tokens_)
    lexical_diversity = (len(set(text2)) / len(text2)) * 100
    print(f"Лексическое разнообразие книги: {lexical_diversity}%")
```

```python
#3
class Tokenizer:
    def split_word(self, text):
        tokens = []
        current_token = ''
        for char in text:
            if char.isalpha():
                current_token += char
            else:
                if current_token:
                    tokens.append(current_token)
                current_token = ''
        if current_token:
            tokens.append(current_token)
        return tokens

txt = 'Сосны обступали тропу плотно, и, хотя истыканное их верхушками небо светилось голубым, в лесу было сумрачно.'
tokenizer = Tokenizer()
tokens = tokenizer.split_word(txt)
print(tokens)
```
