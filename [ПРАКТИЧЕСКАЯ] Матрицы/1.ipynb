```python
class Matrix:
    def __init__(self, matrix: list):
        self.matrix: list = matrix

    def __mul__(self, other):
        new_matrix = []
        if isinstance(other, Matrix):
            col_matrix = []
            for i in range(0, len(self.matrix)):
                for j in range(0, len(self.matrix[0])):
                    col_matrix.append(self.matrix[i][j] * other.matrix[i][j])
                new_matrix.append(col_matrix)
                col_matrix = []
        if isinstance(other, int | float):
            col_matrix = []
            for i in range(0, len(self.matrix)):
                for j in range(0, len(self.matrix[0])):
                    col_matrix.append(self.matrix[i][j] * other)
                new_matrix.append(col_matrix)
                col_matrix = []
        return new_matrix

    def __truediv__(self, other):
        new_matrix = []
        if isinstance(other, Matrix):
            col_matrix = []
            for i in range(0, len(self.matrix)):
                for j in range(0, len(self.matrix[0])):
                    col_matrix.append(self.matrix[i][j] / other.matrix[i][j])
                new_matrix.append(col_matrix)
                col_matrix = []
        if isinstance(other, int | float):
            col_matrix = []
            for i in range(0, len(self.matrix)):
                for j in range(0, len(self.matrix[0])):
                    col_matrix.append(self.matrix[i][j] / other)
                new_matrix.append(col_matrix)
                col_matrix = []
        return new_matrix

    def __add__(self, other):
        new_matrix = []
        if isinstance(other, Matrix):
            col_matrix = []
            for i in range(0, len(self.matrix)):
                for j in range(0, len(self.matrix[0])):
                    col_matrix.append(self.matrix[i][j] + other.matrix[i][j])
                new_matrix.append(col_matrix)
                col_matrix = []
        if isinstance(other, int | float):
            col_matrix = []
            for i in range(0, len(self.matrix)):
                for j in range(0, len(self.matrix[0])):
                    col_matrix.append(self.matrix[i][j] + other)
                new_matrix.append(col_matrix)
                col_matrix = []
        return new_matrix

    def __sub__(self, other):
        new_matrix = []
        if isinstance(other, Matrix):
            col_matrix = []
            for i in range(0, len(self.matrix)):
                for j in range(0, len(self.matrix[0])):
                    col_matrix.append(self.matrix[i][j] - other.matrix[i][j])
                new_matrix.append(col_matrix)
                col_matrix = []
        if isinstance(other, int | float):
            col_matrix = []
            for i in range(0, len(self.matrix)):
                for j in range(0, len(self.matrix[0])):
                    col_matrix.append(self.matrix[i][j] - other)
                new_matrix.append(col_matrix)
                col_matrix = []
        return new_matrix

    def T(self):
        new_matrix = []
        col_matrix = []
        for j in range(0, len(self.matrix[0])):
            for i in range(0, len(self.matrix)):
                col_matrix.append(self.matrix[i][j])
            new_matrix.append(col_matrix)
            col_matrix = []
        return new_matrix

    def __matmul__(self, other):
        new_matrix = []
        for i in range(len(self.matrix)):
            row = []
            for j in range(len(other.matrix[0])):
                element = 0
                for k in range(len(other.matrix)):
                    element += self.matrix[i][k] * other.matrix[k][j]
                row.append(element)
            new_matrix.append(row)
        return new_matrix

    def __str__(self):
        return f"{self.matrix}"
```
