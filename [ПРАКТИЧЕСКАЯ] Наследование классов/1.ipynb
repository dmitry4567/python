```python
class Character:
    def __init__(self, strength: float, dexterity: float, intelligence: float):
        self.strength: float = strength
        self.dexterity: float = dexterity
        self.intelligence: float = intelligence
        self.speed: float = self.calculate_speed()
        self.damage: float = self.calculate_damage()

    def calculate_speed(self) -> float:
        return self.dexterity * 2

    def calculate_damage(self) -> float:
        return self.strength + self.intelligence

    def trade(self):
        pass

    def wait(self):
        pass

    def think(self):
        pass

    def move(self):
        pass
```

```python
class Hero(Character):
    def __init__(self, strength: float, dexterity: float, intelligence: float, has_weapon: bool):
        super().__init__(strength, dexterity, intelligence)
        self.has_weapon: bool = has_weapon

    def buy_weapon(self):
        self.has_weapon = True
        print("Оружие куплено")

    def sell_weapon(self):
        self.has_weapon = False
        print("Оружие продано")


class Villain(Character):
    def __init__(self, strength: float, dexterity: float, intelligence: float, has_home: bool):
        super().__init__(strength, dexterity, intelligence)
        self.has_home: bool = has_home

    def buy_home(self):
        if (not self.has_home):
            self.has_home = True
            print("Дом куплен")
        else:
            print("Уже есть дом")

    def sell_home(self):
        if (self.has_home):
            self.has_home = False
            print("Дом продан")
        else:
            print("Дома нет")
```

```python
class Policeman(Hero):
    def __init__(self, strength: float, dexterity: float, intelligence: float, has_weapon: bool, armor: float):
        super().__init__(strength, dexterity, intelligence, has_weapon)
        self.armor: float = armor

    def add_armor(self, count: float):
        self.armor += count
        print("Броня увеличена на", count, "и сейчас она равна", self.armor)

    def reduce_armor(self, count: float):
        self.armor -= count
        print("Броня уменьшена на", count, "и сейчас она равна", self.armor)


class Thief(Villain):
    def __init__(self, strength: float, dexterity: float, intelligence: float, has_home: bool, color_home: str, house_price: int):
        super().__init__(strength, dexterity, intelligence, has_home)
        self.color_home: str = color_home
        self.house_price: int = house_price

    def change_color_home(self, new_color: str):
        if (self.has_home):
            self.color_home = new_color
            print("Цвет дома изменен на", new_color)
        else:
            print("Нет дома чтобы изменить у него цвет")

    def increase_house_price(self, count: int):
        if (self.has_home):
            self.house_price += count
            print("Стоимость дома повысилась на", count, "и сейчас она равна", self.house_price)
        else:
            print("Нет дома чтобы повысить его цену")

    def reduce_house_price(self, count: int):
        if (self.has_home):
            self.house_price -= count
            print("Стоимость дома понизилась на", count, "и сейчас она равна", self.house_price)
        else:
            print("Нет дома чтобы понизить его цену")
```

```python
policeman_1 = Policeman(strength=10, dexterity=20,
                        intelligence=30, has_weapon=False, armor=10)
policeman_2 = Policeman(strength=10, dexterity=20,
                        intelligence=30, has_weapon=False, armor=10)
thief_1 = Thief(strength=50, dexterity=20, intelligence=40,
                has_home=False, color_home="зеленый", house_price=10000)
thief_2 = Thief(strength=50, dexterity=20, intelligence=40,
                has_home=True, color_home="зеленый", house_price=10000)

policeman_1.add_armor(10)
policeman_2.reduce_armor(5)
policeman_1.buy_weapon()
policeman_1.sell_weapon()

thief_2.buy_home()
thief_2.change_color_home("желтый")
thief_1.change_color_home("красный")
thief_1.increase_house_price(15000)
```
